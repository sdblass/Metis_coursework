```python
def get_soup_driver(base_url): # Retrieves soup from website using Selenium, stores html in repo
#     driver = webdriver.Chrome()
    driver.get(base_url)
    time.sleep(10 + random.uniform(-1,1))
    r = driver.page_source
#     repository[base_url] = r
#     pickling_on = open(fig_path + "zip_info.pickle","wb")
#     pickle.dump(repository, pickling_on)
#     pickling_on.close()    
#     driver.quit()
    return bs(r, 'html.parser')  
```

```python
def get_demog_info(zipcode, repository):
    base_url = f'https://www.unitedstateszipcodes.org/{zipcode}/'
    soup = get_soup_driver(base_url)
    table = soup.find_all('table')
    dfs = pd.read_html(str(table))
    repository[zipcode] = dfs
    pickling_on = open(fig_path + "zip_info_tables.pickle","wb")
    pickle.dump(repository, pickling_on)
    pickling_on.close()  

#     display(df)
#     driver.quit()
    return dfs
```

```python
def clean_zips(zipp):
    zipp = str(zipp)
    if len(zipp) < 5:
        zipp = '0' + zipp
    if len(zipp) < 4:
        zipp = '0' + zipp
    return zipp
    
```

```python
def run_get_demog_info(path, repository, counter=0):
    if counter == 5: return
    global driver
    try:
        driver = webdriver.Chrome()
        zip_df = pd.read_csv(fig_path + 'zips.csv', index_col=0)
        zip_df.dropna(inplace=True)
        zip_list = zip_df['Zip Code']
        zip_list = zip_list.apply(lambda x: clean_zips(x))
        for zipcode in zip_list:
            try:
                if zipcode not in list(repository.keys()):
                    get_demog_info(zipcode, repository)
                else:
                    continue

            except ValueError as err:
                print(err)
                continue
    except Exception as e:
        print(counter)
        print(e)
        counter += 1
        driver.quit()
        time.sleep(5)
        run_get_demog_info(path, repository, counter)
        
```

```python
def extract_data(zipcode, search_term, column, dataset):
    dfs = dataset[zipcode]
    for df in dfs:
#         display(df)
#         df = pd.DataFrame(df)
        search_list = list(df.iloc[:,0])
#         print(search_list)
        try:
            index = search_list.index(search_term)
#             print(index)
            return df.iloc[index, column]
        except ValueError:
            continue
         
    
```

```python
def assemble_zip_demog(zipcode, search_params, dataset):
    results = {}
    for key in search_params:
        try:
            results[key] = extract_data(zipcode, key, search_params[key], emp)
        except Exception as e:
            print(e)
            continue
    results = pd.Series(results)
    results.name = zipcode
    return results
```

```python
def create_zip_demog_df(zip_list, search_params, dataset):
    results_list = []
    for zipcode in zip_list:
        result = assemble_zip_demog(zipcode, search_params, dataset)
        results_list += [result]
    return pd.concat(results_list, axis=1).transpose()
```
